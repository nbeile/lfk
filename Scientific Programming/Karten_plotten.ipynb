```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader

fig = plt.figure(figsize=(10,15))
ax = plt.axes(projection=ccrs.Robinson())

shpfilename = shpreader.natural_earth(resolution="110m", category="cultural", name="admin_0_countries")

reader = shpreader.Reader(shpfilename)
print(reader.geometries())

for geometry in reader.geometries():
    ax.add_geometries(geometry, crs=ccrs.PlateCarree(), edgecolor=(0.7, 0.7, 0.7))

ax.set_global()
ax.coastlines(resolution="110m")
ax.gridlines()

plt.show
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader

fig = plt.figure(figsize=(10,15))
ax = plt.axes(projection=ccrs.Robinson())

shpfilename = shpreader.natural_earth(resolution="110m", category="cultural", name="admin_0_countries")

reader = shpreader.Reader(shpfilename)
print(reader.geometries())

for country in reader.records():
    name = country.attributes["name"]
    if name == "Germany":
        ax.add_geometries(
            country.geometry, 
            crs=ccrs.PlateCarree(), 
            facecolor=(1, 0, 0),
            edgecolor=(0.7, 0.7, 0.7),
        )
    else:
            ax.add_geometries(
                country.geometry, 
                crs=ccrs.PlateCarree(), 
                edgecolor=(0.7, 0.7, 0.7)
            )

ax.set_global()
ax.coastlines(resolution="110m")
ax.gridlines()

plt.show
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader

fig = plt.figure(figsize=(8,30))
ax = plt.axes(projection=ccrs.PlateCarree())

shpfilename = shpreader.natural_earth(resolution="10m", category="cultural", name="admin_0_countries")

reader = shpreader.Reader(shpfilename)
print(reader.geometries())

for country in reader.records():
    name = ""
    if "NAME" in country.attributes:
        name = country.attributes["NAME"]
    elif "name" in country.attributes:
        name = country.attributes["name"]
    if name == "Germany":
        ax.add_geometries(
            country.geometry, 
            crs=ccrs.PlateCarree(), 
            facecolor=(1, 0, 0),
            edgecolor=(0.7, 0.7, 0.7),
        )
    else:
            ax.add_geometries(
                country.geometry, 
                crs=ccrs.PlateCarree(), 
                edgecolor=(0.7, 0.7, 0.7)
            )
            
            
x1 = 5.229
y1 = 15.952
x2 = 46.438
y2 = 56.097
ax.set_extent([x1, y1, x2, y2])
ax.coastlines(resolution="10m")
ax.gridlines()
ax.set_aspect(1.4)

plt.show
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
import cartopy.feature as cfeature

fig = plt.figure(figsize=(8,20))
ax = plt.axes(projection=ccrs.Robinson())

shpfilename = shpreader.natural_earth(resolution="110m", category="cultural", name="admin_0_countries")

ax.set_global()
ax.coastlines(resolution="110m")

#add_feature nur für geringe Auflösungen (gesamte Welt)
ax.add_feature(cfeature.OCEAN, zorder=0)
ax.add_feature(cfeature.LAND, zorder=0,edgecolor="black")
ax.add_feature(cfeature.BORDERS, zorder=0, edgecolor="gray")

ax.gridlines()
plt.show
```
